```python
def get_images_fct(DATA_PATH,IMAGE_PATH,PATH, df_data):
    '''
    Returns a list of images
    '''
    PATH = DATA_PATH+IMAGE_PATH+PATH
    im_names = df_data.filename.unique()
    filenames = [os.path.join(PATH, name) + '.jpg' for name in im_names]
    images = skimage.io.imread_collection(filenames)
    print('Number of images: ', len(images))
    return images
```

## Models

```python
target_size=(40,40)
```

### Linear NN

```python
lin= Sequential()

lin.add(InputLayer(input_shape=(target_size[0], target_size[1],1), name='Input_layer'))
lin.add(Flatten(name='Flattening_layer'))
lin.add(Dense((800), activation='relu', name='First_linear'))
lin.add(Dense((100), activation='relu', name='Second_linear'))
lin.add(Dropout(0.5, name='First_dropout', ))
lin.add(Dense((25), activation='relu', name='Third_linear'))
lin.add(Dropout(0.5, name='Second_dropout'))
lin.add(Dense((1), activation='sigmoid', name='Output_layer'))

lin.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

lin.summary()
```

### CNN

```python
model = Sequential()

#feature extraction part
model.add(Conv2D(input_shape=(target_size[0],target_size[1], 1), filters=64, 
                 kernel_size=(3,3), activation='relu', name='First_Conv'))
model.add(MaxPooling2D(pool_size=(2, 2), name='First_MaxPool'))

model.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu', 
                 name='Second_Conv'))
model.add(MaxPooling2D(pool_size=(2, 2), name='Second_MaxPool'))

model.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu',
         name='Third_Conv'))
model.add(MaxPooling2D(pool_size=(2, 2), name='Third_MaxPool'))

#classification part
_,b,c,d=model.output_shape #hardcode the reshape for cv.dnn issues
model.add(Reshape(target_shape=[b*c*d], name='Flattening')) 
model.add(Dense(64, activation='relu', name='First_linear'))
model.add(Dropout(0.5, name='Dropout'))
model.add(Dense(1, activation='sigmoid', name='Output_layer'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
model.load_weights('Cnn_gray.h5')
```

### Sliding window

```python
def sliding_window(image, stepSize, windowSize):
    # slide a window across the image
    for y in range(0, image.shape[0], stepSize):
        for x in range(0, image.shape[1], stepSize):
            # yield the current window
            yield (x, y, image[y:y + windowSize[1], x:x + windowSize[0]])
```

```python
def draw_rect(df,img_idx, ax):
    """
    ...
    """
    i=df[df['filename']==df.loc[img_idx,'filename']]
   
    for j in i.index:
            rect = patches.Rectangle((i.iloc[j,1], i.iloc[j,2]),
                                     i.iloc[j,3]-i.iloc[j,1],
                                     i.iloc[j,4]-i.iloc[j,2],
                        linewidth=1,edgecolor='r',facecolor='none')
            ax.add_patch(rect)
    return ax
```

```python
def normalise_colors_values(image):
    '''
    nomalizes images distribution across RGB channels to match target and returns the normalized image
    '''
    
    new_im = np.ones_like(image).astype(float)
    target=[236.8, 240.36, 238.6]
      
    for i in range(3):
        mean = image[:,:,i].mean()
        new_im[:,:,i] = ((target[i]/mean) * image[:,:,i] )
        new_im = new_im.astype(int)
        new_im[new_im > 255] = 255
    return new_im

def rgb2gray(rgb):
  '''
  converts image to RGB (opencv.color had weird behaviour with norm_image) with
  custom weighting for each color.
  '''
  r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]

  r_r = 0.05
  r_g = 0.9
  r_b = 0.05

  #3 randoms floats summing to 1
  if 0==1:
      r_r = random.uniform(0,1)
      r_g = random.uniform(0,1)
      r_b = random.uniform(0,1)

      sum_tot = r_r + r_g + r_b

      r_r = r_r/sum_tot
      r_g = r_g/sum_tot
      r_b = r_b/sum_tot

  gray = r_r * r + r_g * g + r_b * b

  return gray.astype(int)
```

```python
#Size of the sliding windows (a bit bigger than a varroa which is approximately a square of 32x32)
(winW, winH) = (40, 40)
overlap=20
img=cv2.imread(DATA_PATH+'/images/train/'+df_train.loc[0, 'filename']+'.jpg')

#Train images processing
probas=np.float32(np.zeros_like(img[:,:,0]))
mask=np.float32(np.zeros_like(img[:,:,0])) #how many time a pixels gets updates

# loop over the sliding window
for (x, y, window) in sliding_window(img, stepSize=overlap, windowSize=(winW, winH)):
    
    # if the window does not meet our desired window size, ignore it
    if window.shape[0] != winH or window.shape[1] != winW:
        continue
    else:
        window=normalise_colors_values(window)
        window=rgb2gray(window)/255
        window=window[np.newaxis, ..., np.newaxis]
        probas[y:y+overlap,x:x+overlap] = probas[y:y+overlap,x:x+overlap] + model.predict_proba(window)[0][0]
        mask[y:y+overlap,x:x+overlap] +=  mask[y:y+overlap,x:x+overlap] + 1

probas=probas/mask #get the mean of the predictions for each pixel
```

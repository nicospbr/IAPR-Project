```python
import skimage.io
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline
import shutil
import os
```

```python
import os

label = 'test'
annotation_path = os.path.join('..', 'data', 'project-data', 'annotations', label)
img_path = os.path.join('..', 'data', 'project-data', 'images', label)
ref_file = os.path.join('..', 'data', 'project-data', label + '.txt')
```

```python
with open(ref_file, 'r') as f:
    ids = f.readlines()
ids = [i.replace('\n', '') for i in ids]
ids[:10]
```

```python
import xml.etree.ElementTree as ET

def parse_file(filename, compact=False):
    """ Parse a PASCAL VOC xml file """
    tree = ET.parse(filename)
    objects = []
    for obj in tree.findall('object'):
        obj_struct = {}
        obj_struct['name'] = obj.find('name').text
        bbox = obj.find('bndbox')
        obj_struct['bbox'] = [int(float(bbox.find('xmin').text)),
                              int(float(bbox.find('ymin').text)),
                              int(float(bbox.find('xmax').text))-int(float(bbox.find('xmin').text)),
                              int(float(bbox.find('ymax').text))-int(float(bbox.find('ymin').text))]
        if not compact:
            objects.append(obj_struct)
        else:
            objects.append(obj_struct['bbox'])

    return objects
```

```python
data_xmls = [parse_file(os.path.join(annotation_path, i) + '.xml', compact=True) for i in ids]
data_jpgs = [os.path.join(img_path, i) + '.jpg' for i in ids]
```

```python
id_display = 2
print('Varroa in {}'.format(data_jpgs[id_display]))
data_xmls[id_display]
```
